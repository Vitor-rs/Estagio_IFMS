```python
import pandas as pd

# Lendo o arquivo
df = pd.read_excel(r"C:\Users\70089478100\OneDrive\Documentos\AREAS\TI\Análise de dados\Professor Leucivaldo\Dados Pesquisa COMPLETA.xlsx")

# Vamos ver as primeiras linhas da coluna específica
print("Primeiras linhas da coluna:")
print(df['Salário mínimo vigente - R$'].head())

# Verificar o tipo de dados atual
print("\nTipo de dados da coluna:")
print(df['Salário mínimo vigente - R$'].dtype)

# Agora vamos fazer a conversão
def convert_br_number(value):
    try:
        if isinstance(value, str):
            # Remove espaços e substitui vírgula por ponto
            value = value.strip().replace(',', '.')
            return float(value)
        elif isinstance(value, (int, float)):
            return float(value)
        else:
            return None
    except:
        print(f"Erro ao converter o valor: {value}")
        return None

# Aplicando a conversão
df['Salário mínimo vigente - R$'] = df['Salário mínimo vigente - R$'].apply(convert_br_number)

# Verificando o resultado
print("\nDepois da conversão - primeiras linhas:")
print(df['Salário mínimo vigente - R$'].head())

# Verificando se há valores não convertidos (None)
nulos = df['Salário mínimo vigente - R$'].isnull().sum()
print(f"\nQuantidade de valores não convertidos: {nulos}")

# Verificando o novo tipo de dados
print("\nNovo tipo de dados da coluna:")
print(df['Salário mínimo vigente - R$'].dtype)

```

```python
import pandas as pd

# Lendo o arquivo
df = pd.read_excel('C:\\Users\\70089478100\\OneDrive\Documentos\\AREAS\\TI\Análise de dados\\Professor Leucivaldo\\Dados Pesquisa COMPLETA.xlsx')

# Converter a coluna de data
def convert_date(value):
    try:
        if isinstance(value, str):
            # Separar ano e mês
            year, month = value.split('.')
            # Criar data no formato americano (YYYY-MM-DD)
            # Como os dias não são relevantes, usamos dia 01
            return f"{year}-{month.zfill(2)}-01"
        return value
    except:
        print(f"Erro ao converter a data: {value}")
        return None

# Aplicar a conversão
df['Data'] = df['Data'].apply(convert_date)

# Converter para datetime
df['Data'] = pd.to_datetime(df['Data'])

# Se você quiser apenas ano e mês no resultado final
df['Data'] = df['Data'].dt.to_period('M')

# Verificar o resultado
print("\nPrimeiras linhas após conversão:")
print(df['Data'].head())

# Salvar o arquivo
df.to_excel(r'C:\\Users\\70089478100\\OneDrive\Documentos\\AREAS\\TI\Análise de dados\\Professor Leucivaldo\\Dados Pesquisa COMPLETA.xlsx', 
            index=False)
```

```python
import pandas as pd

# Lendo o arquivo
df = pd.read_excel(r"C:\Users\vitor\OneDrive\Documentos\AREAS\TI\Análise de dados\Professor Leucivaldo\Dados Pesquisa COMPLETA.xlsx")

# Renomear as colunas de data para maior clareza
# Verifique se as colunas existem antes de renomear
if 'Data' in df.columns and 'Data.1' in df.columns:
    df = df.rename(columns={
        'Data': 'Data_Inicial',
        'Data.1': 'Data_Commodities'
    })
else:
    print("As colunas 'Data' e/ou 'Data.1' não existem no dataframe.")

# Função para converter números com vírgula para float
def convert_br_number(value):
    try:
        if isinstance(value, str):
            return float(value.replace(',', '.'))
        return float(value) if pd.notna(value) else None
    except:
        return None

# Função para converter datas
def convert_to_timestamp(value):
    try:
        if isinstance(value, str):
            year, month = value.split('.')
            return pd.Timestamp(f"{year}-{month.zfill(2)}-01")
        return value
    except:
        return None

# Verifique se as colunas 'Data_Inicial' e 'Data_Commodities' existem antes de aplicar a conversão
if 'Data_Inicial' in df.columns:
    df['Data_Inicial'] = df['Data_Inicial'].apply(convert_to_timestamp)
else:
    print("A coluna 'Data_Inicial' não existe no dataframe.")

if 'Data_Commodities' in df.columns:
    df['Data_Commodities'] = df['Data_Commodities'].apply(convert_to_timestamp)
else:
    print("A coluna 'Data_Commodities' não existe no dataframe.")

# Converter colunas numéricas
colunas_numericas = [
    'Salário mínimo vigente - R$',
    'IPCA - geral - índice (dez, 1993 = 100) Mensal',
    'Câmbio',
    'PIB - US$ (bilhões)',
    'SELIC Mensal',
    'Soja',
    'Arroz',
    'Milho',
    'Trigo Paraná'
]

for coluna in colunas_numericas:
    if coluna in df.columns:
        df[coluna] = df[coluna].apply(convert_br_number)
    else:
        print(f"A coluna '{coluna}' não existe no dataframe.")

# Verificar os tipos de dados resultantes
print("\nTipos de dados após conversão:")
print(df.dtypes)

# Verificar algumas linhas para confirmar a conversão
print("\nPrimeiras linhas após conversão:")
print(df.head())

# Verificar se há valores não convertidos em cada coluna
print("\nQuantidade de valores nulos por coluna:")
print(df.isnull().sum())

# Salvar o arquivo convertido
df.to_excel(r"C:\Users\vitor\OneDrive\Documentos\AREAS\TI\Análise de dados\Professor Leucivaldo\Dados Pesquisa COMPLETA_convertido.xlsx", 
            index=False)
```
